```python
import os
import time
import pandas as pd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from geopy.geocoders import Nominatim
# define a geocoder
geolocator = Nominatim(user_agent="sample app",timeout=3)
```

```python
data = []
for index, row in df2.iterrows():
    lat = row['lat']
    lon = row['lon']
    name = row['name']
    cap = row['capacity']
    data.append([lat, lon, name, cap])  
    
dataset = pd.DataFrame(data, columns=['latitude','longitude','name','capacity'])
dataset.head(4)
```

```python
# visualize the bike stations location and status
locations = dataset[['latitude', 'longitude']]
locationlist = locations.values.tolist()
map = folium.Map(location=[43.662643, -79.395689], zoom_start=14)
for point in range(0, len(locationlist)):
    if locationlist[point]== [0.0, 0.0]: continue
    popup_message = "Capacity="+str(dataset['capacity'][point])+"\n\n"+dataset['name'][point]
    folium.Marker(locationlist[point], popup=popup_message).add_to(map)
map
```

```python
# creating and rendering a cluster map
map = folium.Map(location=[43.662643, -79.395689], zoom_start=11)

marker_cluster = MarkerCluster().add_to(map)

for point in range(0, len(locationlist)):
    popup_message = "Capacity="+str(dataset['capacity'][point])+"\n\n"+dataset['name'][point]
    folium.Marker(locationlist[point], popup=popup_message).add_to(marker_cluster)
map
```

```python
# Hexagonal Binning map
import plotly.figure_factory as ff

fig = ff.create_hexbin_mapbox(
    data_frame=dataset, 
    lat="latitude", 
    lon="longitude", 
    nx_hexagon=15, 
    opacity=0.5,
    min_count=1,
    labels={'color':'#SharedBikeStation'},
    color_continuous_scale='turbo'
    )

fig.update_layout(mapbox_style="open-street-map")
fig.show()
```

```python
ridership_data = pd.read_csv('ridership.csv')
ridership_data
```

```python
data = []
for index, row in ridership_data.iterrows():
    start_time = row['Start Time']
    end_time = row['End Time']
    startSID = row['Start Station Id']
    endSID = row['End Station Id']
    BID = row['Bike Id']
    # remove rows where endSID=NaN
    if (pd.isna(endSID)):
        continue
    data.append([start_time, end_time, startSID, endSID, BID])  
    
ridership_df = pd.DataFrame(data, columns=['start_time','end_time','startSID','endSID','BID'])
ridership_df
```

```python
# 150 rows removed
# save as new csv
ridership_df.to_csv('ridership.csv')
```
